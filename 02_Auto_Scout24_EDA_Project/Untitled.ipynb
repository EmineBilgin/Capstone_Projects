# CENTRAL LIMIT THEOREM

## Sample Mean for a Uniform Distribution

```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
random.seed(54312)
```

```python
sample_size = 30
sim_num = 1000
```

```python
sample_list = [[random.randint(0,100) for i in range(sample_size)] for j in range(sim_num)]
```

```python
mean_list = [sum(i)/sample_size for i in sample_list]
```

```python
plt.hist(mean_list, bins=100, density=True, color='r')
plt.grid()
mu = 50
sigma = math.sqrt(((100**2)/12)) / (math.sqrt(sample_size))
x = np.linspace(mu- 4*sigma, mu + 4*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

# CONFIDENCE INTERVAL

```python
sample_size = 30
sample_list = [random.randint(0,10) for i in range(sample_size)]

sample_mean = np.mean(sample_list)
n = len(sample_list)
cl = .95
std = 1
```

```python
(1-cl)/2 + cl
```

```python
critic_value = stats.norm.ppf((1-cl)/2 + cl)
```

```python
lower_limit = sample_mean - (critic_value * (std/math.sqrt(n))) 
upper_limit = sample_mean + (critic_value * (std/math.sqrt(n))) 
```

```python
stats.norm.interval(cl, loc=sample_mean, scale=std/math.sqrt(n))
```
