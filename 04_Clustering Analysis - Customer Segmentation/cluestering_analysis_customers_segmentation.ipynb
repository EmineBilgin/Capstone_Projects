## 3. Cluster Analysis

The main purpose of this project is to perform [cluster analysis](https://en.wikipedia.org/wiki/Cluster_analysis#:~:text=Cluster%20analysis%20or%20clustering%20is,in%20other%20groups%20(clusters).) with the [K-Means](https://towardsdatascience.com/understanding-k-means-clustering-in-machine-learning-6a6e67336aa1) algorithm. 

We can perform many [cluster analysis](http://www.stat.columbia.edu/~madigan/W2025/notes/clustering.pdf) using different variables. If we use a maximum of two variables for each cluster analysis, we can identify cluster labels more clearly.

First, the K-Means algorithm expects you to determine the number of clusters (*n_clusters*). We can determine the optimal number of clusters for each cluster analysis in various ways. In this case, We use the [Elbow Method](https://en.wikipedia.org/wiki/Elbow_method_(clustering).

Finally, different information is obtained in each analysis. Therefore, different labeling should be done as a result of each cluster analysis. 

Labeling example: 

- **Normal Customers**  -- An Average consumer in terms of spending and Annual Income
- **Spender Customers** --  Annual Income is less but spending high, so can also be treated as potential target customer.

### Clustering for whole dataset with K-Means

#### *i. Preprocessing our dataset to clustering*

While CustomerID doesn't contain meaningful information for the clustering purposes we have dropped this column. On the other hand, as Gender consists of cathegorical data and k-means algorithm isn't directly applicable to categorical variables (because Euclidean distance function isn't really meaningful for discrete variables) we may drop this column too. However, Gender might contain valuable insights for our model we also have an option to use group_by method to see Gender effect.

Let's have both a gender based and general data frames and apply all process to all data frames (mall_g, mall_m, mall_f).

```python
mall_g = mall.drop(columns = ['Gender'], axis=1)
mall_g.head(3)
```

```python
mall_f = mall[mall.Gender==1].drop(columns = ['Gender'], axis=1)
mall_f.head(3)
```

```python
mall_m = mall[mall.Gender==0].drop(columns = ['Gender'], axis=1)
mall_m.head(3)
```

Now that clustering our dataset has features with different magnitudes and we would like to have an equally distribution, we are going to scale our data frames in order to answer these needs.

```python
mall_g = StandardScaler().fit_transform(mall_g)
mall_g[0:5]
```

```python
mall_f = StandardScaler().fit_transform(mall_f)
mall_f[0:5]
```

```python
mall_m = StandardScaler().fit_transform(mall_m)
mall_m[0:5]
```

#### *ii. Determine optimal number of clusters*

We are going to use Elbow Method to determine optimum K value.

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

# Instantiate the clustering model and visualizer
visualizer = KElbowVisualizer(KMeans(), k=(2,15))
plt.figure(figsize=(8,4)).patch.set_facecolor('xkcd:white')
visualizer.fit(mall_g)       # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```python
# Instantiate the clustering model and visualizer
visualizer = KElbowVisualizer(KMeans(), k=(2,15), metric='calinski_harabasz', timings=False)
plt.figure(figsize=(8,4)).patch.set_facecolor('xkcd:white')
visualizer.fit(mall_g)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```python
# Instantiate the clustering model and visualizer
visualizer = KElbowVisualizer(KMeans(), k=(2,15), metric='calinski_harabasz', timings=False)
plt.figure(figsize=(8,4)).patch.set_facecolor('xkcd:white')
visualizer.fit(mall_f)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```python
# Instantiate the clustering model and visualizer
visualizer = KElbowVisualizer(KMeans(), k=(2,15), metric='calinski_harabasz', timings=False)
plt.figure(figsize=(8,4)).patch.set_facecolor('xkcd:white')
visualizer.fit(mall_m)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

#### *iii. Apply K Means*

We found best K-values for mall_g (all customers), mall_f (female customers) and mall_m (male customers). We are going to apply these K-values to our models for each data frames.

```python
k_means = KMeans(init = "k-means++", n_clusters = 6, n_init = 12)
k_means.fit(mall_g)
```

```python
mall_g = pd.DataFrame(mall_g).rename(columns = {0:'Age', 1:'Annual Income (k$)', 2:'Spending Score (1-100)'})
mall_g['customer_type'] = k_means.labels_
```

```python
mall_g.head(3)
```

```python
k_means_f = KMeans(init = "k-means++", n_clusters = 6, n_init = 12)
k_means_f.fit(mall_f)
```

```python
mall_f = pd.DataFrame(mall_f).rename(columns = {0:'Age', 1:'Annual Income (k$)', 2:'Spending Score (1-100)'})
mall_f['customer_type'] = k_means_f.labels_
```

```python
mall_f.head(3)
```

```python
k_means_m = KMeans(init = "k-means++", n_clusters = 4, n_init = 12)
k_means_m.fit(mall_m)
```

```python
mall_m = pd.DataFrame(mall_m).rename(columns = {0:'Age', 1:'Annual Income (k$)', 2:'Spending Score (1-100)'})
mall_m['customer_type'] = k_means_m.labels_
```

```python
mall_m.head(3)
```

#### *iv. Visualizing Labeled Clusters*

```python
f, ax = plt.subplots(3, 2, sharey=True,figsize=(16,12))
ax[0,0].set_title('K Means General "Annual Income vs. Spending Score"')
ax[0,0].scatter(mall_g['Annual Income (k$)'],mall_g['Spending Score (1-100)'],c=mall_g['customer_type'],cmap='rainbow')
ax[0,1].set_title('K Means General "Age vs. Spending Score"')
ax[0,1].scatter(mall_g['Age'],mall_g['Spending Score (1-100)'],c=mall_g['customer_type'],cmap='rainbow')

ax[1,0].set_title('Female K Means "Annual Income vs. Spending Score"')
ax[1,0].scatter(mall_f['Annual Income (k$)'],mall_f['Spending Score (1-100)'],c=mall_f['customer_type'],cmap='rainbow')
ax[1,1].set_title('Female K Means "Age vs. Spending Score"')
ax[1,1].scatter(mall_f['Age'],mall_f['Spending Score (1-100)'],c=mall_f['customer_type'],cmap='rainbow')

ax[2,0].set_title('Male K Means "Annual Income vs. Spending Score"')
ax[2,0].scatter(mall_m['Annual Income (k$)'],mall_m['Spending Score (1-100)'],c=mall_m['customer_type'],cmap='rainbow')
ax[2,1].set_title('Male K Means "Age vs. Spending Score"')
ax[2,1].scatter(mall_m['Age'],mall_m['Spending Score (1-100)'],c=mall_m['customer_type'],cmap='rainbow')
```

Visually having 5 clusters instead of 6 could be more reasonable. However we can not decide just from visual aspect. Let's ignore age and make whole process with a so called simplified data frames :

### Clustering based on Annual Income and Spending Score

#### *i. Creating a simplified dataset*

```python
mall_is =  pd.DataFrame(StandardScaler().fit_transform(mall.drop(columns = ['Gender', 'Age'], axis=1))).rename(columns = {0:'Annual Income (k$)', 1:'Spending Score (1-100)'})
mall_is.head()
```

```python
plt.figure(figsize=(10, 8))
plt.title('K Means General "Annual Income vs. Spending Score"')
plt.scatter(mall_is['Annual Income (k$)'],mall_is['Spending Score (1-100)'],c=mall_is['Customer_type'],cmap='rainbow')
```

### Hierarchical Clustering

In this section we are going to use an other clustering algorithm and try to double check our findings with K-means.

#### *i. Determine optimal number of clusters using Dendogram*

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial import distance_matrix
from sklearn.cluster import AgglomerativeClustering 
```

Since this is a double check effort we are going to check all data frame configurations we already checked in previpus section. To do so we are going to build Dendograms for all these data frames and try to form clustering for each case.

```python
hc_complete_g = linkage(mall_g.drop(columns='customer_type', axis=1), "complete")
hc_ward_g = linkage(mall_g.drop(columns='customer_type', axis=1), "ward")

hc_complete_f = linkage(mall_f.drop(columns='customer_type', axis=1), "complete")
hc_ward_f = linkage(mall_f.drop(columns='customer_type', axis=1), "ward")

hc_complete_m = linkage(mall_m.drop(columns='customer_type', axis=1), "complete")
hc_ward_m = linkage(mall_m.drop(columns='customer_type', axis=1), "ward")

hc_complete_as = linkage(mall_as.drop(columns='Customer_type', axis=1), "complete")
hc_ward_as = linkage(mall_as.drop(columns='Customer_type', axis=1), "ward")

hc_complete_is = linkage(mall_is.drop(columns='Customer_type', axis=1), "complete")
hc_ward_is = linkage(mall_is.drop(columns='Customer_type', axis=1), "ward")
```

We have formed linkage matrixes regarding both ward and complete methodologies. We are going to plot our dendograms for each cases and state the n_cluster numbers for each case regarding this dendrograms.

```python
plt.figure(figsize=(20,20))
plt.grid(b=None)
plt.subplot(521).set_title('Dendrogram_Ward (General)')
plt.subplot(521)
dendrogram(hc_ward_g, leaf_font_size = 7)
plt.subplot(522).set_title('Dendrogram_Complete (General)')
plt.subplot(522)
dendrogram(hc_complete_g, leaf_font_size = 7)

plt.subplot(523).set_title('Dendrogram_Ward (Female)')
plt.subplot(523)
dendrogram(hc_ward_f, leaf_font_size = 7)
plt.subplot(524).set_title('Dendrogram_Complete (Female)')
plt.subplot(524)
dendrogram(hc_complete_f, leaf_font_size = 7)

plt.subplot(525).set_title('Dendrogram_Ward (Male)')
plt.subplot(525)
dendrogram(hc_ward_m, leaf_font_size = 7)
plt.subplot(526).set_title('Dendrogram_Complete (Male)')
plt.subplot(526)
dendrogram(hc_complete_m, leaf_font_size = 7)

plt.subplot(527).set_title('Dendrogram_Ward (Age vs. Spending)')
plt.subplot(527)
dendrogram(hc_ward_as, leaf_font_size = 7)
plt.subplot(528).set_title('Dendrogram_Complete (Age vs. Spending)')
plt.subplot(528)
dendrogram(hc_complete_as, leaf_font_size = 7)

plt.subplot(529).set_title('Dendrogram_Ward (Income vs. Spending)')
plt.subplot(529)
dendrogram(hc_ward_is, leaf_font_size = 7)
plt.subplot(5,2,10).set_title('Dendogram_Complete (Income vs. Spending)')
plt.subplot(5,2,10)
dendrogram(hc_complete_is, leaf_font_size = 7)

plt.show();
```

#### ii. *Apply Agglomerative Clustering*

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agg_model_g = AgglomerativeClustering(n_clusters=6, linkage = 'complete')
agg_model_as = AgglomerativeClustering(n_clusters=4, linkage = 'complete')
agg_model_is = AgglomerativeClustering(n_clusters=5, linkage = 'complete')
```

```python
mall_agg_g = mall_g.drop(columns='customer_type', axis=1)
mall_agg_as = mall_as.drop(columns='Customer_type', axis=1)
mall_agg_is = mall_is.drop(columns='Customer_type', axis=1)
```

```python
mall_agg_g['customer_type'] = agg_model_g.fit_predict(mall_agg_g)
mall_agg_as['customer_type'] = agg_model_as.fit_predict(mall_agg_as)
mall_agg_is['customer_type'] = agg_model_is.fit_predict(mall_agg_is)
```

```python
f, ax = plt.subplots(2, 2, sharey=True,figsize=(16,12))
ax[0,0].set_title('Agglomerative General "Annual Income vs. Spending Score"')
ax[0,0].scatter(mall_agg_g['Annual Income (k$)'],mall_agg_g['Spending Score (1-100)'],c=mall_agg_g['customer_type'],cmap='rainbow')
ax[0,1].set_title('Agglomerative General "Age vs. Spending Score"')
ax[0,1].scatter(mall_agg_g['Age'],mall_agg_g['Spending Score (1-100)'],c=mall_agg_g['customer_type'],cmap='rainbow')

ax[1,0].set_title('Agglomerative mall_is "Annual Income vs. Spending Score"')
ax[1,0].scatter(mall_agg_is['Annual Income (k$)'],mall_agg_is['Spending Score (1-100)'],c=mall_agg_is['customer_type'],cmap='rainbow')
ax[1,1].set_title('Agglomerative mall_as "Age vs. Spending Score"')
ax[1,1].scatter(mall_agg_as['Age'],mall_agg_as['Spending Score (1-100)'],c=mall_agg_as['customer_type'],cmap='rainbow')
```

### Conclusion

In this project we built both K_means and Hierarchical Clustering models to get valuable intuitions 
out of our data.

We tried to see clusterings regarding designated features via forming various data frames in order to see
the clustering effects.

* Data frame formed with three features 'Annual Income', 'Age' and 'Spending Score' tend to cluster as 6 type of customers.

Last but not least  K_means and Hierarchical Clustering have similar results with 'Annual Income' and 'Spending Score' however
